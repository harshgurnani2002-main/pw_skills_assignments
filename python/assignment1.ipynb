```python
# qyuestion 10 : Write code to demonstrate type casting with list elements.

x=['10','20','30','40']
print("before type casting")
for i in x:
    print(f"value : {i} , type : {type(i)}")
x=[int(i) for i in x]
print("after type casting")
for i in x:
    print(f"value : {i} , type : {type(i)}")
```

```python
# question 14 : Write a program to reverse a string using a while loop.

string=input("enter a string")
rev_string=""
index=len(string)-1
while index>=0:
    rev_string+=string[index]
    index-=1
print(f"reversed string is : {rev_string}")
```

```python
#question 15 : Write a program to calculate the factorial of a number using a while loop.
num=int(input("enter a number"))
factorial=1
while num>0:
    factorial*=num
    num-=1
print(f"factorial is : {factorial}")
```
