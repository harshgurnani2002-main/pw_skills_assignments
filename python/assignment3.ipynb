```python
# question 1 write a python function that takes a list of numbers as input and returns the sum of all the even numbers in the list

def sumofeven(numbers)->float:
    return sum(x for x in numbers if x%2==0) #pythonic one liner

lst=[1,2,3,4,5,6]
print(sumofeven(lst))
```

```python
# question 2 create a python function that accepts a string and returns the reverse of string

def revstr(string:str)->str:
    return string[::-1]

s="python"
print(revstr(s))
```

```python
# question 3 implement a python function that takes a list of integers and returns a new list containing the squares of each number

def retsquare(lst:list)->list:
    l2=[x**2 for x in lst]
    return l2

l1=[1,2,3,4,5,6]
print(retsquare(l1))
```

```python
# question 4 write a python function to check if a given number is prime or not from 1 to 200

def isPrime(n:int)->bool:
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
        else:
            return True

number=17
print(isPrime(number))
```

```python
# question 5 create an iterator class in python that generates a fibonacci sequence up to a specified number of terms

## creating iter of class for fibonacci

class Fibonacci:
    def __init__(self,n:int)->None:
        self.n=n
        self.count=0
        self.a=0
        self.b=1

    def __iter__(self):
        return self
    def __next__(self):
        if self.count>=self.n:
            raise StopIteration

        if self.count==0:
            self.count+=1
            return 0
        elif self.count==1:
            self.count+=1
            return 1

        result=self.a+self.b
        self.a=self.b
        self.b=result
        self.count+=1
        return result


fib=Fibonacci(10)

for num in fib:
    print(num)
```

```python
#question 6 write a generator function in python that yeilds the powers of 2 up to a given exponent


def gen_up_to(n:int)->list:
    for i in range(1,n+1):
        yield  2**i

gen=gen_up_to(10)
for num in gen:
    print(num)
```

```python
# question 7 implement a generator function that reads a file line by line and yields each line as a string

def read_file_line_by_line(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line.strip()
file_path='sample.txt'
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
# question 8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

tuples_list = [
    (101, 89),   # student_id, score
    (102, 72),
    (103, 95),
    (104, 68)
]

sorted_tuples = sorted(tuples_list, key=lambda x: x[1])
print(sorted_tuples)
```

```python
# question 9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celcius_temps=[0,100,40,80]

farenhiet_temps=list(map(lambda c:(c*9/5)+32,celcius_temps))
print(farenhiet_temps)
```

```python
# question 10  Create a Python program that uses `filter()` to remove all the vowels from a given string.



text='hello my name is harsh and i am writing code in python language '

def remove_vowels(input_string: str) -> str:
    vowel='aeiouAeiou'
    result=''.join(filter(lambda c: c not in vowel,input_string))
    return result

filtered_text=remove_vowels(text)

print(filtered_text)
```
